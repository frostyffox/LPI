```python
x = [-2, -1, 1, 2]
print("The absolute value of the vector x is: ", np.abs(x))
print("The exponential of the vector x is:: ", np.exp(x))
print("The logarithm of the vector : ", np.log(np.abs(x)))
```

```python
# Creating an array from a list
#L = [[-5, 6.9, 34, 0],[-2, -1, 1, 2]]
L = [-5, 6.9, 34, 0,-2, -1, 1, 2]
L_array = np.array(L)
L_array.shape #return the dimension of the array in a tuple
```

```python
# Creating an array directly
A = np.random.randint(0,100,[2,3])
B = np.random.randint(0,10,[3,2])
print(A)
print(B)
# Example of a dot product betweeen two matrices
print(np.dot(A,B)) #or np.dot(A@B)
#B = [[-5, 6.9, 34, 0],[-2, -1, 1, 2]]
C = np.random.randint(0,10,[2,5])
print(np.dot(np.dot(A,B),C))
```

```python
# Getting the type of the elements of the array
L_array.dtype
```

```python
# Array Indexing
# accessing the fifth element of a one-dimensionnal arrray
print(L_array[4])
# accessing all elements starting the 3rd
print(L_array[2:])
# accessing the first three elements
print(L_array[:3])
# every second element
print(L_array[::2])
```

```python
# Conditional Indexing
A = np.random.randint(0,100,[2,3])
print(A)
print(A[A>1]) #checking if each element of the matrix is >1 
#the result is an array of type boolean
#the condition is put into the square brackets

#B= A>1
#print(B.dtype)

L_array[(L_array > 2000)]
print((L_array > 2000))
# array([2200, 2172])
L_array[(L_array > 2000) & (L_array < 3000)]
# array([2200, 2172])
```
